Faça um Programa que peça as 4 notas bimestrais e mostre a média.

```
nota1 = int(input("Digite sua primeira nota "))
nota2 = int(input("Digite sua segunda nota "))
nota3 = int(input("Digite sua terceira nota "))
nota4 = int(input("Digite sua quarta nota "))
media = (nota1 + nota2 + nota3 + nota4)/4 
print("A sua media é {0}".format(media))
```

Faça um Programa que peça o raio de um círculo, calcule e mostre sua área.

```
import math
raio = int(input("Digite o valor do raio de um círculo "))
diametro = math.pow(raio,2) * math.pi
print("O seu diametro é {0}".format(round(diametro,2)))
```

Faça um Programa que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.


```
import math
lado = int(input("Digite o valor de um lado do quadrado "))
area = math.pow(lado,2) 
print("A área do quadrado é {0} e o dobro dela é {1}".format(round(area), round(area*2)))
```

Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
C = 5 * ((F-32) / 9).

```
fahr = int(input("Digite o valor da temperatura em graus Fahrenheit "))
celsius  = 5*((fahr-32)/9)
print("A temperatura em celsius é {0}º".format(round(celsius,2)))
```

Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit

```
celsius = int(input("Digite o valor da temperatura em graus Fahrenheit "))
fahr  =  ((celsius*9)/5) + 32
print("A temperatura em celsius é {0}º".format(round(fahr,2)))
```
