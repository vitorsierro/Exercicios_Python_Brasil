```
import pandas as pd
import numpy as np
ano = []
pop = []
excel  = pd.read_csv('https://raw.githubusercontent.com/vitorsierro/csv_and_xls/main/populacao-brasileira.csv')
excel = excel.to_numpy()
for i in range(len(excel)):
    ano.append(excel[i,0])
    pop.append(excel[i,1])
print(ano)
print(pop)
```
