Crie um algoritmo em Python que peça as 4 notas bimestrais e mostre a média.

```
p1 = float(input("Digite sua nota da primeira prova "))
p2 = float(input("Digite sua nota da segunda prova "))
p3 = float(input("Digite sua nota da terceira prova "))
p4 = float(input("Digite sua nota da quarta prova "))
soma = (p1 + p2 + p3 + p4) / 4
print("A media foi {0}".format(soma))
```
