Faça um Programa que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

```
#-*- coding: utf-8 -*-
def quadrado(lado):
  area = lado * lado
  perimetro = lado + lado + lado + lado
  return(area, perimetro)
lado = int(input("Digite o valor do lado do quadrado\n"))
area, perimetro = quadrado(lado)
print("O valor da area é", area)
print("O valor do perimetro é", perimetro)
print("O dobro do valor da area é", area*2)
```

Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
* Para homens: (72.7*h) - 58
* Para mulheres: (62.1*h) - 44.7

```
#-*- coding: utf-8 -*-
alt = int(input("Digite sua altura em metros? \n"))
sexo = int(input("Você é Homem(1) ou Mulher(2)? \n"))
if sexo == 1:
  calc = ((72.7*alt) - 58)
elif sexo == 2:
  calc = ((62.1*alt) - 44.7)
else:
  print("Você digitou um valor errado")

print("Seu peso ideal é", calc)
```

Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).


```
#-*- coding: utf-8 -*-
arq_down = int(input("Digite o tamanho do arquivo que dejesa baixar em MB\n"))
net = int(input("Digite a velocidade de um link de Internet (em Mbps),\n"))
Mb = arq_down * 1000
net = net / 8
calc = Mb / net
calc_mi = calc / 60
print("O valor que vai ser baixado é em Minutos é", calc_mi)
```
